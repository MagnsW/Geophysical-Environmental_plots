```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = './input/curves_90deg/'
filename1 = path + 'mb_90deg_3000m_120km_40m.txt'
```

```python
f = open(filename1)
lines = f.read().splitlines()
f.close()
for i, line in enumerate(lines):
    if i < 30:
        print(i, line)
```

```python
column_names = ['Azimuth', 'Range', 'Depth', 'SEL']
```

```python
shot1 = pd.DataFrame()
```

```python
shot1 = pd.read_csv(filename1, names=column_names, skiprows=1, delim_whitespace=True)
```

```python
shot1
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
#sns.set_palette("Pastel")
```

```python
plt.figure(figsize=(16,8))
plt.axhline(160, color='red', linestyle=':')
plt.axhline(150, color='red', linestyle=':')
plt.axhline(140, color='red', linestyle=':')
plt.axhline(130, color='red', linestyle=':')
plt.xticks(np.arange(200, 125000, step=5000))
plt.ylim(120, 180)
plt.title(filename1)

ax = sns.regplot(x="Range", y="SEL", data=shot1, logx=True)
```
